# Executive Summary

## Project Objective

This project seeks to use data science to understand:
- What are the factors in a job listing that are most important in predicting whether a job posting in the data science domain is above or below the median salary across data science related jobs. 
- What are the factors in a job listing that help to distinguish data scientist jobs from other data science related jobs? 
- What are the factors in a job listing that help to distinguish juinor and senior data science-related jobs?




### Data Source

I used the 'MyCareersFuture' job portal, an initiative by the Singapore Government to match Singaporeans/PRs with potential employers. There is a rich source of information that accompanies every job listing, including salary information thus making this data source suitable for this project. 
-   Company name
-   Job Title
-	Employment Type
-	Seniority
-	Job Category
-	Salary Range
-	Roles & Responsibilities
-	Requirements 



### Methodology

The relevant data was obtained by using an automated software to scour through the job portal for keywords pertaining to data science related jobs openings. Code for the automated crawler can be found in the appendix. 

Subsequently, the relevant data source is split into a training dataset and a testing dataset. A few machine learning models - RandomForestClassifier, Naive Bayes and SVM were used to train the training dataset, and validated using the testing dataset. This process was used to answer each of the questions listed in the project objective. 

Natural Language Processing algorithms were also utilised to extract interesting features from unstructured text such as the job description and job requirements. 



### Findings

***i) What are the factors in a job listing that are most important in predicting whether a job posting in the data science domain is above or below the median salary across data science related jobs.***
- Applying Random Forest to job categorical information - Seniority, Employment Type, Job Category and Job_Tag (different data science keywords that I applied in the job search query) resulted in a good accuracy score of ~0.75; The same score was also obtained by running the model against unstructured Job Requirements text. Based on the Job requirements text, the top 5 features in determining salary above median include - ***Big Data, Maching Learning, Years Experience, 10 years and Project Management***

- However, using the same model to unstructured Roles & Responsibilities text as well as job titles text was less effective with an accuracy score of ~0.61 and ~0.65 respectively. 
- Applying the Naive Bayes method on job categorical information resulted in an accuracy score of 0.75 as well, not different from applying Random Forest.


***ii) What are the factors in a job listing that help to distinguish data scientist jobs from other data science related jobs***
- There was a class imbalance issue as there were many more non-data scientist jobs than data scientist jobs. This would have ot be ideally resolved by using various sampling methods so that differences between the classes will be more pronounced by the features. 
- I first attempted to apply the RandomForest model to distinguish the job classes without addressing the class imbalance. This resulted in an accuracy score of 91%, 3% higher than the baseline of 88%. Based on this, the top 5 features in distinguishing data scientist jobs from other data science related jobs are - ***Data Science, Statistics Applied, Machine Learning, Data Mining and Quantitative Field. ***
- Attempting an oversampling of dataset, and also applying the RandomForest model, the accuracy score improved to 95%. However, this better result might be an indication of overfitting. The top 5 features in this instance in distinguishing data scientist jobs from other data science related jobs are - ***Machine Learning, Data Science, Data Mining, Quantitative Field, Deep Learning***



***iii) What are the factors in a job listing that help to distinguish juinor and senior data science-related jobs?***
- I arbitrarily segregated the dataset by Job seniority categories into junior and senior categories. To ensure that we do not include any misclassified job listings, I included an additional filter for median salary across the entire dataset. For junior roles, take only listings with salaries lower than the median salary, and vice versa for senior roles.
- Using a RandomForest model achieved an accuracy score of ~79%, which is much better than the baseline accuracy of 53%. The top 5 features in this instance in distusing junior and senior data science jobs are: ***10 years, Open Source, Degree Diploma, Big Data, Computer Science and Track Record ***

### 1) EDA and Data Cleaning

```python
# Read scraped output file

jobs = pd.read_csv('190130 Job listings (Final).csv')
print(jobs.shape)
jobs.head()
```

```python
# Check if there are any missing values

jobs.isnull().sum()
```

Let's investigate the missing values in Salary_Frequency

**1.1 EDA and Clean salary columns - Salary Range, Salary Min/Max, Salary Frequency**

```python
# Check undisclosed salary ranges
jobs['Salary_Range'].value_counts()
```

```python
# Check if undisclosed salary ranges correspond with $0 for Min and Max salary

jobs[jobs['Salary_Range'] == 'Salary undisclosed']
```

It appears that jobs with undisclosed salary have a maximum salary but salary frequency is not disclosed. Since there are no salary values tagged to these 106 jobs, let's drop them.

```python
# Drop rows with no salary information

jobs.dropna(axis = 0, subset = ['Salary_Frequency'], inplace = True)
```

```python
jobs.shape
```

```python
# Convert salary_Min and salary_Max columns to numeric

jobs['Salary_Max'] = jobs['Salary_Max'].apply(lambda x: x.replace('$', ''))
jobs['Salary_Max'] = jobs['Salary_Max'].apply(lambda x: x.replace(',', ''))

jobs['Salary_Min'] = jobs['Salary_Min'].apply(lambda x: x.replace('$', ''))
jobs['Salary_Min'] = jobs['Salary_Min'].apply(lambda x: x.replace(',', ''))

# Convert salary columns to integers
jobs['Salary_Max'] = jobs['Salary_Max'].astype(int)
jobs['Salary_Min'] = jobs['Salary_Min'].astype(int)


# Convert all annual salaries to monthly salaries. Assumption of 2 month bonus in annual package, so 12+2 mths bonus
jobs['Salary_Max'] = np.where(jobs['Salary_Frequency'] == 'Annually', jobs['Salary_Max']/14, jobs['Salary_Max'].astype(int))
jobs['Salary_Min'] = np.where(jobs['Salary_Frequency'] == 'Annually', jobs['Salary_Min']/14, jobs['Salary_Min'].astype(int))
```

```python
# Change listings with annual salary frequency to monthly after conversion

jobs['Salary_Frequency'].replace('Annually', 'Monthly', inplace = True)
```

```python
# Check distribution of Salary_Frequency

jobs['Salary_Frequency'].value_counts()
```

```python
# Check if there are any abnormal values for salary_Max and salary_Min using boxplot

plt.figure(figsize = (18,5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

ax1.set_title("Distribution for Salary_Min")
ax2.set_title("Distribution for Salary_Max")

sns.boxplot(x = jobs['Salary_Min'], ax = ax1)
sns.boxplot(x = jobs['Salary_Max'], ax = ax2); 
```

It appears that there are some outliers of listings with Salary_Min > 40000. Let's take a closer look at these listings. 

```python
# Check listings for salary_min

jobs[jobs['Salary_Min'] > 40000]
```

It appears to be a fat finger mistake that these salaries ranges indicated in these listings are intended to be monthly salaries rather than annual salaries. Let's convert them to monthly salaries.

```python
# Adjust records with salary_min > 40000

jobs['Salary_Min'] = jobs['Salary_Min'].map(lambda x: x/14 if x > 40000 else x)
jobs['Salary_Max'] = jobs['Salary_Max'].map(lambda x: x/14 if x > 40000 else x)
```

```python
# Relook at boxplot again
plt.figure(figsize = (18,5))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

ax1.set_title("Distribution for Salary_Min")
ax2.set_title("Distribution for Salary_Max")

sns.boxplot(x = jobs['Salary_Min'], ax = ax1)
sns.boxplot(x = jobs['Salary_Max'], ax = ax2); 
```

```python
# Create a new salary average column derived from those two columns that we can use as a benchmark salary for each listing

jobs['Salary_Avg'] = (jobs['Salary_Max'] + jobs['Salary_Min'])/2
```

```python
# Let's visualise the salary distribution

bins = np.histogram(np.hstack((jobs['Salary_Min'],jobs['Salary_Avg'],jobs['Salary_Max'])), 
                    bins=25)[1]

plt.figure(figsize = (18,5))

plt.hist(jobs['Salary_Max'],bins, label='Max', alpha=0.4)
plt.hist(jobs['Salary_Avg'],bins, label='Mid', alpha=0.4)
plt.hist(jobs['Salary_Min'],bins, label='Min', alpha=0.6)

plt.title('Salary Distribution', fontsize=16)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Salary', fontsize=14)
plt.ylabel('Counts', fontsize=14)

plt.show()
```

Given that distribution of Salary_Avg is smoother than the Salary_Min and Salary_Max columns, we will use that as the target variable. 

We will address this question as a classification problem of whether a salary is high or low with other salary information provided. We will use the median salary of Salary_Avg as the tipping point of whether a salary is considered high or low. 

```python
median_salary = jobs['Salary_Avg'].median()
median_salary
```

```python
# Apply salary label for each median salary

jobs['Salary_Label'] = jobs['Salary_Avg'].apply(lambda x: 1 if x >= median_salary else 0)
```

```python
# Check distribution of salary labels

jobs['Salary_Label'].value_counts()
```

```python
# Drop the Salary_Range column

jobs.drop(['Salary_Range'], axis = 1, inplace = True)
```

### 2) Modelling

```python
# Create a copy of cleaned up dataset
jobs_model = jobs.copy()
```

**2.1 Model based on Employment_Type, Job_Category, Seniority and Job_Tag (i.e. categorical columns)**

```python
X_21 = jobs_model[['Employment_Type', 'Job_Category', 'Seniority', 'Job_Tag']]
y = jobs_model['Salary_Label']
```

```python
X_21 = pd.get_dummies(X_21)
```

```python
# Train/Test split

X_train21, X_test21, y_train21, y_test21 = train_test_split(X_21, y, test_size = 0.3, random_state = 22, stratify = y)
```

```python
# Try using Random Forest Classifier

rfc21 = RandomForestClassifier(n_estimators = 500, random_state = 23)    

rfc21.fit(X_train21, y_train21)

y_pred21 = rfc21.predict(X_test21)
acc21 = accuracy_score(y_test21, y_pred21)
print('Accuracy Score: ', acc21)
    
cv21 = cross_val_score(rfc21, X_train21, y_train21, cv = 10, n_jobs = -1, scoring = 'accuracy')
print('Cross Validation Score: {:0.3} ± {:0.3}'.format(cv21.mean().round(3), cv21.std().round(3)))
```

```python
# Plot confusion matrix

confusion_mat = pd.crosstab(y_test21, y_pred21,
                            rownames=['Actual'],
                            colnames=['Predicted'],margins=True)
confusion_mat
```

```python
# Print classification report

print(classification_report(y_test21, y_pred21))
```

```python
# Plotting ROC curve for high salary (label 1)
fpr, tpr, thresholds = roc_curve(y_test21, y_pred21)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=[8,8])
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc, linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('ROC for high salary', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```

For a first model, the results look pretty good based off just various job categorical variables.

```python
search_parameters = ['data%20scientist',
                     'data%20analyst',
                     'data%20engineer',
                     'data%20architect',
                     'python',
                     'machine&20learning',
                     'data%20manager',
                     'machine%20learning',
                     'deep%20learning',
                     'data%20science',
                     'big%20data',
                     'data%20management']
```

```python
driver = webdriver.Chrome('/Users/user/chromedriver/chromedriver.exe')
listing_number = 0
job_listings = []

for param in search_parameters:
    url = 'https://www.mycareersfuture.sg/search?search={}&sortBy=new_posting_date&page=0'.format(param)
    driver.get(url)
    sleep(10)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    sleep(3)

    # take total number of results divided by 20, which is the max number of results on each page
    num_of_results = soup.find("span", {'id': 'search-results-message'}).text.split()[0]
    num_of_results = int(num_of_results.replace(',',''))
    num_of_pages = math.ceil(num_of_results/20)
    
    param_formatted = param.replace('%20', " ").title()
    print('Num of pages for {} role: {}'.format(param_formatted, num_of_pages))

    for num in range(0, num_of_pages):
        url = 'https://www.mycareersfuture.sg/search?search={}&sortBy=new_posting_date&page={}'.format(param, num)
        driver.get(url)
        sleep(10)

        html = driver.page_source
        soup = BeautifulSoup(html, 'lxml')
        sleep(3)

        job_card = []
        for row in soup.findAll("div", id = re.compile('job-card')):
            job_card.append(row.get('id'))

        # loop through each job to gather necessary information
        for job in job_card:

            try:
                job_title = soup.find("div", id = job).find("h1", {"name": "job_title"}).text
            except AttributeError:
                job_title = 'Not Available'

            try:
                company = soup.find("div", id = job).find("p", {"name": "company"}).text
            except AttributeError:
                company = 'Not Available'

            try:
                employment_type = soup.find("div", id = job).find("p", {"name": "employment_type"}).text
            except AttributeError:
                employment_type = 'Not Available'

            try:
                seniority = soup.find("div", id = job).find("p", {"name": "seniority"}).text
            except AttributeError:
                seniority = 'Not Available'

            try:
                job_category = soup.find("div", id = job).find("p", {"name": "category"}).text
            except AttributeError:
                job_category = 'Not Available'

            try:
                salary_range = soup.find("div", id = job).find("span", class_ = re.compile("salary_range")).text
            except AttributeError:
                salary_range = 'Not Available'

            try:
                salary_min = salary_range.split('to')[0]
            except IndexError:
                salary_min = 'Not Available'

            try:
                salary_max = salary_range.split('to')[1]
            except IndexError:
                salary_min = 'Not Available'

            try:
                salary_freq = soup.find("div", id = job).find("span", {'class': 'salary_type'}).text
            except AttributeError:
                salary_freq = 'Not Available'

            try:
                URL = soup.find("div", id = job).find("a", href=re.compile("/job/")).get('href')
            except AttributeError:
                URL = 'Not Available'
                
            role_type = param_formatted
            
            job_site = 'MyCareersFuture'

            job_listings.append([job_title, company, employment_type, seniority, job_category, salary_range, salary_min, salary_max, \
                         salary_freq, URL, role_type])  
            listing_number += 1
            print(listing_number)

        print("Page {} is complete".format(num))
    print("{} related jobs complete".format(param_formatted))
    
    # Store local copy in case of interruption
    df_desc.to_csv('./{}.csv'.format(param_formatted), encoding = 'utf-8', index = False)

driver.close()
print("Crawling is complete.")
```

```python
# Initiating empty dataframe to store basic job-related information
df = pd.DataFrame(data = job_listings, columns=["Job_Title","Company","Employment_Type","Seniority","Job_Category","Salary_Range", 
                                                "Salary_Min","Salary_Max", "Salary_Frequency","URL", "Job_Tag"])
```

### 2) Scrape job description and requirements

```python
# Collecting descriptions and requirements across all listings

driver = webdriver.Chrome('/Users/user/chromedriver/chromedriver.exe')
listing_number = 0
description_list = []

df_desc = pd.DataFrame(data = description_list, 
                       columns=["Desc_content", "Requirements"])

for row in df['URL']:
    url = 'https://www.mycareersfuture.sg{}'.format(row)
    driver.get(url)
    sleep(5)

    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    sleep(5)
    
    try:
        desc_content = soup.find("div", {'id': 'description-content'}).text
    except AttributeError:
        desc_content = 'Not Available'
        
        
    try:
        requirements = soup.find("div", {'id': 'requirements'}).text
    except AttributeError:
        requirements = 'Not Available'
        
    description_list.append([row, desc_content, requirements])  
    listing_number += 1
    print(listing_number)
    if listing_number in [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]:
        df_desc = pd.DataFrame(data = description_list, columns=["row","Desc_content", "Requirements"])
        
        # Store local copy in case of interruption
        df_desc.to_csv('./df_desc {}.csv'.format(listing_number), encoding = 'utf-8', index = False)
driver.close()
print("Crawling is complete.")
```
