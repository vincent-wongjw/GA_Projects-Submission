Write a for-loop that:

### 6.2

How many times did that loop run? If you have N gyms and also N players, how many times would it run as a function of N?

Can you think of a more efficient way to accomplish the same thing? 

(You can write your answer as Markdown text.)

````python
print('The loop ran for {} times.'.format(loop_counter1))
````

The loop ran 20 times given that there are 2 players and 10 gyms. 

In the case of N gyms and N players, the number of loops would be N**2 which is a factor of the number of gyms AND players. 

A more efficient way that involves less loops would be to look at set intersection between the number of gyms that each player has visited against the list of gym locations. This would only involve 2 loops as highlighted in the code below. 


````python
loop_counter2 = 0

for i in players:
    loop_counter2 += 1
    intersect_gyms = list(set(players[i]['gyms_visited']).intersection(set(gym_locations)))
    if len(intersect_gyms) != 0:
        print('{player_name} has visited {gym} {and_last_gym}'.format(
            player_name = players[i]['player_name'],
            gym = intersect_gyms[0] if len(intersect_gyms) == 1 else ', '.join(intersect_gyms[0:-1]),
            and_last_gym = '' if len(intersect_gyms) == 1 else ' and {}'.format(intersect_gyms[-1])
        ))
    else:
        print('{} has not visited any gyms'.format(players[i]['player_name']))
        
print('The loop ran for {} times.'.format(loop_counter2))
````

## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def player_power(players, pokedex, player_id):
    player_power = 0
    for pokemon in players[player_id]['player_pokemon']:
        attack_and_defense = pokedex[pokemon]['attack'] + pokedex[pokemon]['defense']
        pokemon_power = attack_and_defense + pokedex[pokemon]['speed'] + pokedex[pokemon]['hp'] + pokedex[pokemon]['special_attack'] + pokedex[pokemon]['special_defense']
        player_power += pokemon_power

    print("{}'s power is {}".format(players[player_id]['player_name'], player_power))
    return player_power
        
player_power(players, pokedex, 1)
player_power(players, pokedex, 2)
````

While you were putting together the prototype code, your colleagues were preparing a dataset of Pokemon and their attributes. (This was a rush job, so they may have picked some crazy values for some...)

The code below loads information from a comma separated value (csv) file. You need to parse this string into a more useable format. The format of the string is:

- Rows are separated by newline characters: \n
- Columns are separated by commas: ,
- All cells in the csv are double quoted. Ex: "PokedexNumber" is the first cell of the first row.


Using for-loops, create a list of lists where each list within the overall list is a row of the csv/matrix, and each element in that list is a cell in that row. Additional criteria:

1. Quotes are removed from each cell item.
2. Numeric column values are converted to floats.
3. There are some cells that are empty and have no information. For these cells put a -1 value in place.

Your end result is effectively a matrix. Each list in the outer list is a row, and the *j*th elements of list together form the *j*th column, which represents a data attribute. The first three lists in your pokedex list should look like this:

    ['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']
    [1.0, 'Bulbasaur', 'GrassPoison', 318.0, 45.0, 49.0, 49.0, 65.0, 65.0, 45.0]
    [2.0, 'Ivysaur', 'GrassPoison', 405.0, 60.0, 62.0, 63.0, 80.0, 80.0, 60.0]

````python
import csv

pokemon_list = []

pokedex_file = 'pokedex_basic.csv'
with open(pokedex_file, 'r') as f:
    raw_pd = csv.reader(f)
    for line in raw_pd:
        pokemon_list.append(line)

pokemon_list_float = []
    
for row in pokemon_list:
    new_row = []
    for value in row:
        if value.isnumeric() is True:
            value = float(value)
            new_row.append(value)
        elif value == '':
            new_row.append('-1')
        else:
            new_row.append(value)
    pokemon_list_float.append(new_row)


pprint(pokemon_list_float)
````

## 10. Calibrate the frequency of Pokemon

The design team wants you to make the powerful Pokemon rare, and the weaklings more common. How would you set the probability $p_i$ of finding Pokemon *i* each time a player visits a gym?

Write a function that takes in a Pokedex number and returns a value $p_i$ for that character.

Hint: there are many ways you could do this. What do _you_ think makes sense? Start with simplifying assumptions: for example, you could assume that the probabilities of encountering any two Pokemon on one visit to a gym are independent of each other.

````python
# First, we would have to differentiate between strong pokemon and weak pokemon. A simple way to look at this would be to 
# segregate pokemon based on the 'Total' stat, which is the aggregate power for each pokemon in the pokedex.


# The current distribution of the 'Total' across all pokemon

total_scores = [parsed_pokedex[i+1]['Total'] for i in range(len(parsed_pokedex))]
plt.figure(figsize = (8,5))
plt.hist(total_scores, label = 'No. of Pokemon in each bin')
plt.title("'Total' stat distribution across all Pokemon in the Pokedex")
plt.xlabel("Pokemon 'Total' stat distribution")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
````

````python
# Currently, it is assumed that there is an equal probability of encountering pokemon regardless of power
# Assumptions to recalibrate frequency of Pokemon:
#     - probabilities of encountering any two Pokemon on one visit to a gym are independent of each other

#     - The probability of encountering a pokemon should be inversely proportional to the strength of that pokemon

#     - Thus, we could use the reciprocal of 'Total' score of each pokemon, and subsequently the probability of encountering
#         a pokemon would be the reciprocal of that pokemon's 'Total' over the sum of all such scores for all pokemons. 


def pokedex_with_probability(pokedex):

# Gather sum of reciprocal of 'Total' score for each pokemon
    score_list = []
    for pokedex_id in pokedex:
        score = 1 / pokedex[pokedex_id]['Total']
        score_list.append(score)

    total_score = sum(score_list)

# Add to dictionary of each pokemon the probability of encontering in a gym
    for pokedex_id in pokedex:
        score = 1 / pokedex[pokedex_id]['Total']
        pokedex[pokedex_id]['Probability to find in gym'] = score / total_score
        
    return pokedex

# A probability score is now assigned to each pokemon in the pokedex
parsed_pokedex = pokedex_with_probability(parsed_pokedex)
pprint(parsed_pokedex)
````

````python
total_scores = [parsed_pokedex[i+1]['Total'] for i in range(len(parsed_pokedex))]

probability_scores = [parsed_pokedex[i+1]['Probability to find in gym'] for i in range(len(parsed_pokedex))]

# Convert probability score into percentage for easier reference
probability_scores = np.array(probability_scores)*100

plt.figure(figsize = (8,5))
plt.scatter(total_scores, probability_scores, label = 'No. of Pokemon in each bin')
plt.title("Probability of encountering in gym based on 'Total' stat distribution of Pokemon")
plt.xlabel("Pokemon 'Total' stat distribution")
plt.ylabel("% chance of encountering in gym")
plt.legend()
plt.grid()

# In the graph below, the probability of encountering a stronger pokemon becomes lower as the 'Total' stat becomes higher. 
````
